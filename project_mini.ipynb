```python
player_name = input("당신의 이름은 무엇입니까? ")
health = 100
item1 = ""
item2 = ""
item3 = ""

print("이상한 차원문🕳️에 빨려들어가 어디론가 떨어집니다.")
print("""어.. 음..
여긴.. 어디..(어느 이상한 숲에서 깨어남)""")
print(player_name + "님, 던전 생존기에 오신 걸 환영합니다!")
print("아이템:", item1, item2, item3)
print("초기 ❤️체력:", health)
print("")

# 월요일
print("숲에서 낡은 상자를 발견하였습니다.")
print("[월요일] 낡은 상자에서 '⚔️검'을 찾았습니다.")
choice = input("⚔️검을 주우시겠습니까? (예/아니오): ")
if choice == "예":
    item1 = "⚔️검"
    print("⚔️검을 장착했습니다!")
else:
    health = health - 20
    print("⚔️무기가 없어 불안합니다. (❤️체력 -10)")

print("경계 가득한 마음으로 뜬눈으로 하루를 보냅니다.")
print("심한 경계심으로 체력이 감소합니다 (❤️체력 -10)")
health = health - 10
print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")

# 화요일
print("[화요일]아침이 되었습니다. 길에서 빵을 발견했습니다.")
eat_bread = input("빵을 먹겠습니까? (예/아니오): ")
if eat_bread == "예":
    print("🪳곰팡이가 가득한 빵입니다.")
    health -= 10
    print("❤️체력이 감소합니다 (❤️체력 -10)")
else:
    print("🪳곰팡이가 들은 🪳빵을 지나가는 🐈고양이가 먹고 쓰러집니다.☠️")
    health += 10
    print(player_name + "곰팡이가 들은 🥖빵을 피했습니다. 신선한 🥖빵을 제공합니다. (❤️체력 +10)")

print(f"현재 ❤️체력: {health}")

if health <= 0:
    print("❤️체력이 바닥나 쓰러졌습니다. ☠️게임 오버☠️")
else:
    print("")
    
if health > 0:
    print("👾슬라임이 다가옵니다!")
    choice = input("👾슬라임과 싸우시겠습니까? (예/아니오): ")
    if choice == "예":
        if item1 == "⚔️검" or item2 == "⚔️검" or item3 == "⚔️검":
            print("슬라임을 처치했습니다! '작은 💊포션'을 얻었습니다.")
            if item2 == "":
                item2 = "작은 💊포션"
            elif item3 == "":
                item3 = "작은 💊포션"
        else:
            health = health - 20
            print("맨손 전투로 상처를 입었습니다. (❤️체력 -20)")
    else:
        health = health - 20
        print("도망치다 부상을 입었습니다. (❤️체력 -10)")

if health < 0:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
else:
    print("")

print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")


# 수요일
print("[수요일] 앞에 두건을 쓴 사람이 보입니다.")
if health > 0:
    print("수상한 상인이 나타났습니다.")
    if item2 == "작은 💊포션" or item3 == "작은 💊포션":
        choice = input("작은 💊포션을 '회복 💊물약'으로 교환하시겠습니까? (예/아니오): ")
        if choice == "예":
            if item2 == "작은 💊포션":
                item2 = "회복 💊물약"
            else:
                item3 = "회복 💊물약"
            print("💊회복 물약을 얻었습니다!")
    else:
        print("교환할 아이템이 없습니다.")

if health < 0:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
else:
    print("")

print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")

# 목요일
print("[목요일] 지나가는 길에 무언가 🦶발에 걸립니다.")
if health > 0:
    print("🪤덫에 걸렸습니다!")
    if item2 == "회복 💊물약" or item3 == "회복 💊물약":
        choice = input("회복 물약을 사용하시겠습니까? (예/아니오): ")
        if choice == "예":
            if item2 == "회복 💊물약":
                item2 = ""
            else:
                item3 = ""
            health = health + 20
            print("회복했습니다! (❤️체력 +20)")
        else:
            health = health - 20
            print("회복하지 않아 다쳤습니다. (❤️체력 -20)")
    else:
        health = health - 20
        print("회복 물약이 없어 다쳤습니다. (❤️체력 -20)")

if health < 0:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
else:
    print("")

print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")

# 금요일
print("[금요일] 좁은 🌋동굴이 보입니다.")
if health > 0:
    print("좁은 길을 지납니다.")
    choice = input("조심히 걷겠습니까? (예/아니오): ")
    if choice == "예":
        print("무사히 지나갔습니다.")
    else:
        health = health - 10
        print("발을 헛디뎌 발목을 다쳤습니다. (❤️체력 -10)")

if health < 0:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
else:
    print("")
    
print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")

# 토요일
print("[토요일] 좁은 골목을 지나오니 커다란 물체가 보입니다.")
if health > 0:
    print("👾보스를 만났습니다!")
    choice = input("👾보스와 싸우시겠습니까? (예/아니오): ")
    if choice == "예":
        if item1 == "⚔️검":
            print("👾보스를 쓰러뜨리고 '마력 수정💎'을 얻었습니다.")
            if item2 == "":
                item2 = "마력 수정💎"
            elif item3 == "":
                item3 = "마력 수정💎"
        else:
            health = health - 40
            print("👊맨손으로 싸워 큰 피해를 입었습니다. (❤️체력 -30)")
    else:
        health = health - 30
        print("🏃‍➡️도망쳤지만 부상을 입었습니다. (❤️체력 -20)")

if health < 0:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
else:
    print("")

print("아이템:", item1, item2, item3)
print("현재 ❤️체력:", health)
print("")

# 일요일
if health > 0:
    print("[일요일] 🌋던전 출구를 발견했습니다.")
    if item2 == "마력 수정💎" or item3 == "마력 수정💎":
        print("마력 수정💎으로 봉인을 풀고 탈출에 성공했습니다!")
        print("🎉 생존 성공! 현실 🌏세계로 돌아갑니다.,", player_name + "님!")
    else:
        print("당신은 살아남았지만 🪦동굴에 같이게 되었습니다.")
        print("생존은 성공!, 탈출 실패!")
    print("최종 ❤️체력:", health)
    print("최종 아이템:", item1, item2, item3)
else:
    print("❤️체력이 바닥났습니다. ☠️게임 오버☠️.")
```
